Scraping Data from Website + Pandas Dataframe

```python
from bs4 import BeautifulSoup
import requests 
```

```python
# Define the URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
# Set headers to identify the scraper (avoids being blocked)
headers = {
    "User-Agent": "EducationalScraper/1.0 (contact: your_email@example.com)"
}

page = requests.get(url,headers=headers) # Fetch the page content
soup = BeautifulSoup(page.text, 'html') # Parse the HTML content with BeautifulSoup
```

```python
first_table = soup.find_all('table')[0]
```

```python
# Get the first row headers
header_row = first_table.find("tr")
World_titles = header_row.find_all("th")
```

```python
# Extract clean text
World_table_titles = [ title.text.strip() for title in World_titles]
print(World_table_titles)
```

```python
import pandas as pd
```

```python
# Create an empty DataFrame with the table headers
df = pd.DataFrame (columns = World_table_titles)
```

```python
# Get all rows of the table
column_data = first_table.find_all('tr')
```

```python
# Loop through table rows (skip header) and add data to DataFrame
for row in column_data [1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    
    length = len(df)
    df.loc [length] = individual_row_data
```
